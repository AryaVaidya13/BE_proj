```python
# ðŸ§ª NLP Research Assistant â€” Auto Experiment Runner
# This notebook dynamically executes experiments based on experiment_plan.json

import json
import pandas as pd
from datasets import load_dataset
from transformers import pipeline

# ---------- 1. Load experiment plan ----------
with open("data/processed/experiment_plan.json", "r") as f:
    plan = json.load(f)["experiments"]

print(f"âœ… Loaded {len(plan)} experiments")

results = []

# ---------- 2. Loop through each experiment ----------
for exp in plan:
    title = exp.get("paper_title", "Untitled")
    dataset_name = exp.get("Datasets used", "imdb")
    model_name = exp.get("Models / methods used", "distilbert-base-uncased-finetuned-sst-2-english")
    metric = exp.get("Evaluation metrics", "accuracy")

    print(f"\nðŸŽ¯ Running experiment for: {title}")
    print(f"ðŸ“š Dataset: {dataset_name} | ðŸ§  Model: {model_name}")

    try:
        # Load dataset (using fallback if not found)
        dataset = load_dataset(dataset_name, split="test[:5]") if dataset_name else load_dataset("imdb", split="test[:5]")

        # Detect task type
        task = "text-classification"
        pipe = pipeline(task, model=model_name)

        sample_text = dataset[0].get("text") or dataset[0].get("sentence") or str(dataset[0])
        output = pipe(sample_text[:512])[0]

        result = {
            "paper_title": title,
            "dataset": dataset_name,
            "model": model_name,
            "metric": metric,
            "prediction": output,
        }
        results.append(result)
        print(f"âœ… Success: {result['prediction']}")

    except Exception as e:
        print(f"âš ï¸ Skipped {model_name} on {dataset_name}: {e}")

# ---------- 3. Save results ----------
df = pd.DataFrame(results)
df.to_json("data/processed/colab_results.json", orient="records", indent=2)
df.to_csv("data/processed/colab_results.csv", index=False)

print("\nâœ… All experiments completed!")
print("ðŸ“ Results saved to: data/processed/colab_results.json and .csv")
```
